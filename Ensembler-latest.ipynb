```python
import pandas as pd
import numpy as np
```

```python
h2o = pd.read_csv("Predictions/h2o_89 89176.csv")
mensembleweighted = pd.read_csv("Predictions/89_86597manualensemble_weighted.csv")

mhighest = pd.read_csv("Predictions/vig_manula_highest89_87982.csv")
autogluon = pd.read_csv("Predictions/autogluon89_84440.csv")
catboost_single=pd.read_csv("Predictions/89_84074catboost.csv")
flaml=pd.read_csv("Predictions/89_84897flaml.csv")
tabnet = pd.read_csv("Predictions/88160_TabNet.csv")
mljar=pd.read_csv("Predictions/mljar_89_88173.csv")
stackedreg=pd.read_csv("Predictions/89_87988stackregression.csv")
catboost_deep=pd.read_csv("Predictions/89_84catboostdeep2.csv")
tabFT=pd.read_csv("Predictions/89_76028tabft.csv")
tpot=pd.read_csv("Predictions/89_83612tpot.csv")
stack_mljar=pd.read_csv("Predictions/stack-mljar.csv")
stack_flaml=pd.read_csv("Predictions/stack-flamel.csv")
stack_flaml_adv=pd.read_csv("Predictions/stack-flamel-a.csv")
stacked_new_new=pd.read_csv("Predictions/89_85290stack.csv")
full_ensemble=pd.read_csv("Predictions\z-rise88_.csv")
manual_wt_new=pd.read_csv("Predictions\manual-weighted-rise.csv")
stack_new=pd.read_csv("Predictions\89_86726stack.csv")
```

```python
weight = {'h2o' : 0.6, 
          'mensembleweighted' : 0.35, 
          'tabnet': 0.45, 
          'mhighest' : 0.55,
          'autogluon' : 0.25, 
          'catboost_deep' : 0.32, 
          'stack_flaml_adv' : 0.1,
          'mljar' : -0.05, 
          'stack_flaml' : -0.25, 
          'stacked_new_new' : -0.15, 
          'stackedreg' : -0.05, 
          'tpot' : -0.15,
          'flaml' : -0.1, 
          'tabFT' : -0.2, 
          'catboost_single' : -0.55,
          'full_ensemble' : -0.02,
          'manual_wt_new' : 0.05,
          'stack_new' : 0.05,
          'stack_mljar' : -0.01}
```

```python
df3 = pd.DataFrame({
    "id": h2o['id'].values,
    "efficiency": (
        weight['stack_mljar'] * stack_mljar['efficiency'].values + 
        weight['stack_new'] * stack_new['efficiency'].values + 
        weight['manual_wt_new'] * manual_wt_new['efficiency'].values + 
        weight['full_ensemble'] * full_ensemble['efficiency'].values + 
        weight['h2o'] * h2o['efficiency'].values + 
        weight['mensembleweighted'] * mensembleweighted['efficiency'].values + #Weighted on ensemble - Catboost,XGBoost,RandomForest,LGBM,ET
        weight['tabnet'] * tabnet['efficiency'].values + 
        weight['mhighest'] * mhighest['efficiency'].values + #Stack Regressor 1 - Catboost,XGBoost,RandomForest,LGBM,ET
        weight['autogluon'] * autogluon['efficiency'].values + 
        weight['catboost_deep'] * catboost_deep['efficiency'].values + 
        weight['stack_flaml_adv'] * stack_flaml_adv['efficiency'].values + 
        weight['mljar'] * mljar['efficiency'].values + 
        weight['stack_flaml'] * stack_flaml['efficiency'].values +
        weight['stacked_new_new'] * stacked_new_new['efficiency'].values + #Stack Regressor – Many Models
        weight['stackedreg'] * stackedreg['efficiency'].values + ##Stack Regressor 3
        weight['tpot'] * tpot['efficiency'].values +
        weight['flaml'] * flaml['efficiency'].values +
        weight['tabFT'] * tabFT['efficiency'].values +
        weight['catboost_single'] * catboost_single['efficiency'].values
    ) / sum(weight.values())
})
```
