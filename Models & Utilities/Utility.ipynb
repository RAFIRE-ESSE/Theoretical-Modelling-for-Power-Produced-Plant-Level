```python
Date_Time = numpy.array(train[['datetime', 
                               'celulas_ctin03_cc_03_1_ir_cel_1', 
                               'celulas_ctin08_cc_08_1_ir_cel_1', 
                               'celulas_ctin03_cc_03_1_ir_cel_2', 
                               'celulas_ctin08_cc_08_1_ir_cel_2',
                               'meteorolgicas_em_03_02_t_dlogger', 
                               'meteorolgicas_em_08_01_t_dlogger']])

T_error = (numpy.array(train[['ppc_p_tot']])-numpy.array(train[['ttr_potenciaproducible']])).reshape(-1)
neer_error = numpy.array([list(T)+[E] for E, T in zip(T_error, Date_Time) if E > -0.1 and E < 0.1])

pandas.DataFrame({'datetime' : numpy.array([i[0] for i in neer_error]),
                  'celulas_ctin03_cc_03_1_ir_cel_1' : numpy.array([float(i[1]) for i in neer_error]),
                  'celulas_ctin08_cc_08_1_ir_cel_1' : numpy.array([float(i[2]) for i in neer_error]),
                  'celulas_ctin03_cc_03_1_ir_cel_2' : numpy.array([float(i[3]) for i in neer_error]),
                  'celulas_ctin08_cc_08_1_ir_cel_2' : numpy.array([float(i[4]) for i in neer_error]),
                  'meteorolgicas_em_03_02_t_dlogger' : numpy.array([float(i[5]) for i in neer_error]),
                  'meteorolgicas_em_08_01_t_dlogger' : numpy.array([float(i[6]) for i in neer_error]),
                  'T_error' : numpy.array([float(i[7]) for i in neer_error])})
```

```python
#average
average = {'celulas_ctin03_cc_03_1_ir_cel_1' : numpy.average(numpy.nan_to_num(numpy.array([float(i[1]) for i in neer_error]), nan=0)),
           'celulas_ctin03_cc_03_1_ir_cel_2' : numpy.average(numpy.nan_to_num(numpy.array([float(i[3]) for i in neer_error]), nan=0)),
           'celulas_ctin08_cc_08_1_ir_cel_1' : numpy.average(numpy.nan_to_num(numpy.array([float(i[2]) for i in neer_error]), nan=0)),
           'celulas_ctin08_cc_08_1_ir_cel_2' : numpy.average(numpy.nan_to_num(numpy.array([float(i[4]) for i in neer_error]), nan=0)),
           'meteorolgicas_em_03_02_t_dlogger' : numpy.average(numpy.nan_to_num(numpy.array([float(i[5]) for i in neer_error]), nan=0)),
           'meteorolgicas_em_08_01_t_dlogger' : numpy.average(numpy.nan_to_num(numpy.array([float(i[6]) for i in neer_error]), nan=0)),
           'T_error' : numpy.average(numpy.nan_to_num(numpy.array([float(i[7]) for i in neer_error]), nan=0))}

for i in average.keys():
    print(f'{i} : {average[i]}')
```
